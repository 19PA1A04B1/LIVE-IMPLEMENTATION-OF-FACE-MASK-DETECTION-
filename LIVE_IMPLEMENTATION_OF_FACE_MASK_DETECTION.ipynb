```python
import cv2
```

```python
image = cv2.imread('Akki.jpg')
assert not isinstance(image,type(None)), 'image not found'
```

```python
image.shape
```

```python
image[0]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(image)
```

```python
haar_data=cv2.CascadeClassifier('data.xml')
```

```python
print(haar_data.detectMultiScale(image))
```

```python
while True:
    faces=haar_data.detectMultiScale(image)
    for x,y,w,h in faces:
        cv2.rectangle(image, (x,y), (x+w, y+h), (255,0,255), 4)
    cv2.imshow('result',image)
    if cv2.waitKey(2)==27:
        break
cv2.destroyAllWindows()
```

```python
capture=cv2.VideoCapture(0)
data1=[]
while True:
    flag, image=capture.read()
    if flag:
        faces=haar_data.detectMultiScale(image)
        for x,y,w,h in faces:
            cv2.rectangle(image, (x,y), (x+w, y+h), (255,0,255), 4)
            face=image[y:y+h ,x:x+w, :]
            face=cv2.resize(face,(50,50))
            print(len(data1))
            if len(data1)<200:
                data1.append(face)
        cv2.imshow('result',image)
        if cv2.waitKey(2)==27 or len(data1)>=200:
            break
capture.release()
cv2.destroyAllWindows()            
```

```python
import numpy as np
```

```python
np.save("without_mask.npy",data1)
```
