```python
#importing opencv as cv2
import cv2 
```

```python
#haar data classification to read the faces of data
#Cascade classifier to classify the images
haar_data=cv2.CascadeClassifier('data.xml') 
```

```python
print(haar_data.detectMultiScale(image))
```

```python
while True:
    faces=haar_data.detectMultiScale(image) #to detect thr faces of image
    for x,y,w,h in faces:
        cv2.rectangle(image, (x,y), (x+w, y+h), (255,0,255), 4)#to create the reactangle frame on faces
    cv2.imshow('result',image)
    if cv2.waitKey(2)==27:#27-ASCII value for escape
        break
cv2.destroyAllWindows()#destroy all the wiondows opened by opencv
```

```python
capture=cv2.VideoCapture(0)#to initialize the capture
data1=[]#data to append the faces
while True:
    flag, image=capture.read()#flag to open the single frame
    if flag:
        faces=haar_data.detectMultiScale(image)
        for x,y,w,h in faces:
            cv2.rectangle(image, (x,y), (x+w, y+h), (255,0,255), 4)#create the rectangle to detect trhe faces
            face=image[y:y+h ,x:x+w, :]
            face=cv2.resize(face,(50,50))#to resize the image
            print(len(data1))
            if len(data1)<200:#if the length of data is less than 200 then faces are appended to data
                data1.append(face)
        cv2.imshow('result',image)#to show the resukt of image
        if cv2.waitKey(2)==27 or len(data1)>=200:
            break
capture.release()#to release the camera
cv2.destroyAllWindows()#to destroy all the windows opened by opencv
```
