```python
import numpy as np #numoy to read the npy data
import cv2#import opencv as cv2
```

```python
with_mask=np.load('with_mask.npy')#load the data
without_mask=np.load('without_mask.npy')
```

```python
with_mask.shape
```

```python
without_mask.shape
```

```python
with_mask=with_mask.reshape(200,50*50*3)#reshape the data
without_mask=without_mask.reshape(200,50*50*3)
```

```python
with_mask.shape
```

```python
without_mask.shape
```

```python
X=np.r_[with_mask,without_mask]#load the data to X
```

```python
X.shape#shape of X
```

```python
labels=np.zeros(X.shape[0])#create the labels
```

```python
labels[200:]=1.0
```

```python
names={0:'Mask',1:'No Mask'}#assign names to output
```

```python
from sklearn.svm import SVC #import support vector machine 
from sklearn.metrics import accuracy_score#import accuracy_score to measure the accuracy
```

```python
from sklearn.model_selection import train_test_split#to import train _test_split to split the data
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,labels,test_size=0.25)#crete test size
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.decomposition import PCA #import principal component analysis
```

```python
pca=PCA(n_components=3) #convert 'n' components to '3'
x_train=pca.fit_transform(x_train) #fit and transform the x_train using PCA
```

```python
x_train.shape
```

```python
svm=SVC()
svm.fit(x_train,y_train) #fit(x_train,y_train) using SVM to classify
```

```python
x_test=pca.fit_transform(x_test) #fit and transform the x_test using PCA
y_pred=svm.predict(x_test) #predict the values using SVM
```

```python
accuracy_score(y_test,y_pred) #print accuracy score
```

```python
haar_data=cv2.CascadeClassifier('data.xml')
capture=cv2.VideoCapture(0)
while True:
    flag, image=capture.read()
    if flag:
        faces=haar_data.detectMultiScale(image)
        for x,y,w,h in faces:
            cv2.rectangle(image, (x,y), (x+w, y+h), (255,0,255), 4)
            face=image[y:y+h ,x:x+w, :]
            face=cv2.resize(face,(50,50))#resize the image
            face=face.reshape(1,-1)#reshape the face
            face=pca.transform(face)#transform the face
            pred=svm.predict(face)[0]#pedict the faces using svm classifier prediction
            n=names[int(pred)]#convert names into int
            print(n)#print mask or no mask
        cv2.imshow('result',image)
        if cv2.waitKey(2)==27:#27 ASCII value of escape
            break
capture.release()#to release the camera
cv2.destroyAllWindows() #to close all windows opened by opencv
```
